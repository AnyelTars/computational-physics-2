```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.025
k = 0.025

x = np.arange(0,1+h, h)
t = np.arange(0,0.1+k, k)

print(t)
```

```python
# BCs
boundaryConditions = [0,0]
initialConditions = np.sin(np.pi*x)

# Matrix T

n = len(x)
m = len(t)

T = np.zeros((n,m))

# All time points, bcs
T[0,:] = boundaryConditions[0]
T[-1,:] = boundaryConditions[1]

# Init. cond.
T[:,0] = initialConditions

# COlumn time, row position
T.round(3)
```

```python
factor = k/h**2

#outer loop


for j in range(1, m):
    for i in range(1,n-1):
        #print(i,j)
        
        T[i,j] = factor*T[i-1, j-1] + (1 - 2*factor)*T[i, j-1] + factor*T[i+1, j-1]
        
T.round(3)
```

```python
plt.plot(T)
plt.legend(t)

plt.xlabel("position")
plt.ylabel("temperature")

```

```python
plt.plot(x,T[:,0])
plt.legend(t)

plt.xlabel("position")
plt.ylabel("temperature")
```

```python
R = np.linspace(1, 0, m)
B = np.linspace(0, 1, m)
G = 0


for j in range(m):
    plt.plot(x, T[:,j], color = [R[j], G, B[j]]) 

plt.legend([f't = {value}s' for value in t])
plt.xlabel("position [m]")
plt.ylabel("temperature [$\degree$ C]")
```
